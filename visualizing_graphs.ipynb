# Starting with an empty graph

```python
import networkx as nx

G = nx.Graph()

if not G.nodes() and not G.edges():
    print("Graph is indeed empty!")
else:
    print("Oops, this graph is not empty!")
```

# Adding some edges and nodes

```python
import networkx as nx

def printGraph(G):
    print("This graph has the following nodes: ")
    print(G.nodes())
    print("This graph has the following edges: ")
    print(G.edges())

G = nx.Graph()

G.add_node("1")  # add single node
G.add_nodes_from(["2", "3"]) # add list of nodes
printGraph(G)

print("------------------")

G.add_edge("1", "3")  # add edge between nodes already in the graph
G.add_edge("4", "5")  # add edge between nodes not in the graph
printGraph(G)
```

# Pretty printing the graph

```python
import networkx as nx
import matplotlib.pyplot as plt

def prettyPrintGraph(G):
    nx.draw(G)
    plt.savefig("figures/simple_graph.png")
    plt.show()


G = nx.Graph()
G.add_node("2")
G.add_edges_from([("1", "3"), ("4", "5")])
prettyPrintGraph(G)
```

# Let's try a cycle

```python
import networkx as nx
import matplotlib.pyplot as plt

def prettyPrintGraph(G):
    nx.draw(G)
    plt.savefig("figures/cycle.png")
    plt.show()


G = nx.Graph()
G.add_edges_from([("1", "2"), ("2", "3"), ("3", "4"), ("4", "5"), ("5", "1")])
prettyPrintGraph(G)
```

# Now a path

```python
import networkx as nx
import matplotlib.pyplot as plt

def prettyPrintGraph(G):
    nx.draw(G)
    plt.savefig("figures/" + str(len(G)) + "_nodes_path.png")
    plt.suptitle("figures/" + str(len(G)) + " Nodes Path")
    plt.show()

G = nx.path_graph(6)
prettyPrintGraph(G)

H = nx.path_graph(100)
prettyPrintGraph(H)
```
